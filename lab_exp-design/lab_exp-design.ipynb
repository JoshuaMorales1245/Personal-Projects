### Part 3.5: Analysis Simpson's Paradox

What we saw in the previous puzzle is actually crazy! Overall, men had a higher admission rate....yet when we looked at Major A only, we saw that the women actually had a higher admission rate. *This trend continues among most majors.* When we stratify by major, we see that the females actually had a higher admission rate than the males.  Run the following cell to see all of the majors *(you'll learn how to use `df.pivot_table` in an upcoming MicroProject)*:

````python
# Run this cell to see all of the majors:
# (You'll learn how to use `df.pivot_table` in an upcoming MicroProject!)
summary = berkeley.pivot_table(columns=["Sex", "Admission"], index="Major", aggfunc="count")
summary.loc[:, ("Year", "F", "% Accepted")] = summary.loc[:, ("Year", "F", "Accepted")] / ( summary.loc[:, ("Year", "F", "Accepted")] + summary.loc[:, ("Year", "F", "Rejected")] )
summary.loc[:, ("Year", "M", "% Accepted")] = summary.loc[:, ("Year", "M", "Accepted")] / ( summary.loc[:, ("Year", "M", "Accepted")] + summary.loc[:, ("Year", "M", "Rejected")] )
summary
````

This phenomenon is called **Simpson's Paradox**. It's an extreme case of confounding because we see the opposite results when we stratify. Simpson's Paradox can occur in Observational Studies. 

In other words, the overall percentages in two separate groups can be misleading because of a confounder. In this case, the confounder was the major. Males happened to be applying to the easier majors more often than females during the fall of 1973.

Once the confounder was controlled for with stratification, we saw that the overall effect reversed.
