Load this dataset into a DataFrame called `df`!

````python
import pandas as pd
df=pd.read_csv("http://waf.cs.illinois.edu/discovery/football.csv")
df
````

### Part 1.1: Calculate the Total Score

In a game of football, two teams play against each other to score points.  Add a new column to the DataFrame called `TotalScore` that adds the `IlliniScore` and `OpponentScore` together:

````python
# Add a `TotalScore` column to the DataFrame:
df["TotalScore"] = df["IlliniScore"] + df["OpponentScore"]
df
````

### Part 1.2: The Craziest Game of Illini Football

Let's consider the "craziest" game of Illini Football the game where the **most total points** were scored.

Create a DataFrame `df_craziest` that contains just one row of the very craziest game:

````python
df_craziest = df.loc[[df["TotalScore"].idxmax()]]
df_craziest
````

### Part 1.3: Find the Worst Game in Illini Football

Let's consider the "worst" game to be the game where the Illini lost by the most points (this is not the game where the opponent scored the most points, but the game where their score had the biggest difference). Create a DataFrame `df_worst` that contains the row(s) with the greatest score difference.

You may want to create a new column to store the "spread" between the the games!

There might be more than one row with the same score difference, so you need to add the use `keep='all'` as an parameter to `df.nlargest()` or `df.nsmallest()` to not drop the duplicates.

````python
df["Spread"] = df["OpponentScore"] - df["IlliniScore"]
df_worst = df.nlargest(n=1, columns = "Spread", keep= "all")
df_worst
````

### Part 1.4: The Best Homecoming Game

````python
df_homecoming = df[df["Note"] == "Homecoming"]
df_homecoming
df_homecoming["IlliniSpread"] = df_homecoming["IlliniScore"] - df_homecoming["OpponentScore"]
df_homecoming_best = df_homecoming.loc[[df_homecoming["IlliniSpread"].idxmax()]]
df_homecoming_best
````

## Part 2: Using Functions for Calculating Values

One of the most useful ways to use functions in Python is to **apply** the function to every row of a DataFrame to generate a new value.  You will write functions of increasing complexity and apply them to the Illini football dataset:

### Part 2.1: Create the Simplest Function

Write a function called `three` that accepts one parameter called `row` and **always** returns `-3`.  You should be able to do this in just 2 lines of code:

````python
# Write the function three:
def three(row):
    return -3
````

To **apply** a function across every row in your DataFrame, you will use the syntax:
```py
df.apply( <function name ONLY>, axis=1 )
```

For example, the following code applies the function `three` to all of the rows in our Illini Football DataFrame:
```
df.apply(three, axis=1)
```

This is the ONLY time in this course we will not put parentheses after the function name!

Run the code below to see the result of `three` on every index in your DataFrame:

````python
df.apply(three, axis=1)
````

We want to save this value into our DataFrame, so we'll create a new column named `score` that will store the result of the apply operation.  We've provided you this code for this first one.  When you run the code:
- Note that there's a new column named `score` on the far-right hand side.
- Note that the value is always `-3`, since that's the return value of your function.

````python
# Creates a new column named `score` that will store the results
# of applying the `three` function across every single row individually:
df["score"] = df.apply(three, axis=1)
df
````

### Part 2.2: Using the `row` parameter

In your function, you had one parameter named `row`.  When calling a function through `apply`, the current row that is being evaluated by the `apply` function will be sent as the `row`.  This will allow you to have access to the column values for the row just like you access a column in a DataFrame.

For example, when inside of your function, you can do the following:
- `row["Opponent"]` will contain the name of the opponent (ex: `"Purdue"`, `"Nebraska"`, etc)
- `row["IlliniScore"]` will contain the score of the Illini (ex: `45`, `23`, etc)
- ...and so on, for all the columns.

Write a function named `calculateTotalScore` that accepts one parameter called `row` and **always** returns the total score of the row by adding the `IlliniScore` value to the `OpponentScore` value.  You should be able to this with one line to define the function and one line as a return statement:

````python
# Write the function calculateTotalScore:
def calculateTotalScore(row):
    return row["IlliniScore"] + row["OpponentScore"]
````

Now, create a new column named `CalculatedTotalScore` that stores the result of applying the `calculateTotalScore` function to the DataFrame:

````python
# Creates a new column named `CalculatedTotalScore` that will store the results
# of applying the `calculateTotalScore` function across every single row individually:
df["CalculatedTotalScore"] = df.apply(calculateTotalScore, axis=1)
df
````

### Part 2.3: Labeling the Game as "Home" or "Away"

Functions become extremely useful when we begin to introduce conditionals to process data based on the value of the row.  In the function below, we are using a new concept -- an `if` statement in Python -- to return if the game is a "Home" game played in Champaign, IL.

The contents inside of the `if` statement are a conditional, just like you used in a DataFrame! 🎉

Run the cell below to define the `homeOrAway` function:

````python
# (You do not need to modify this function, but you MUST run it to define this function.)

def homeOrAway(row):
  # We first check if the data in the `Location` row is "vs.".
  # ...if it is "vs.", we know the game was a "Home" game:
  if row["Location"] == "vs.": return "Home"

  # If we did not return in the previous line of Python code,
  # we know the game was not a "Away" game:
  return "Away"
````

Create a new column named `GameType` that store the result of applying the `homeOrAway` function to the DataFrame:

````python
# Creates a new column named `GameType` that will store the results
# of applying the `homeOrAway` function across every single row individually:
df["GameType"] = df.apply(homeOrAway, axis = 1)
df
````
