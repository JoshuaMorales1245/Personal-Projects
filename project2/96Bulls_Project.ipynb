### Part 2.1: Looking at Measures of Spread and creating new DataFrame
For this part I am going to be taking the **means** and **standard deviations** of important statistics typically used in basketball to determine if a player is preforming good or not. I will compare Michael Jordans stats to the stats of other players on the roster. 


I will be creating a new DataFrame called **"df_over_1000"**  (which contains only players with over 1000 minutes played in order to reduce skewed numbers from outliers when looking at **mean and standard deviation**) for these problems in order to get a more in-depth reading for the team's top performers.

Creating Dataframe "**df_over_1000**":

```python
df_over_1000 = df_no_team_totals[df_no_team_totals["MP"] > 1000].copy()
df_over_1000
```

### Part 2.2: Comparing Total Points Scored in the 95-96 Season Between Top Players

The **MEAN** of **total points** of the DataFrame **df_over_1000**:

```python
mean_points = df_over_1000["PTS"].mean()
mean_points
```

The **STANDARD DEVIATION** of **total points** of the DataFrame:

```python
std_points = df_over_1000["PTS"].std()
std_points
```

The **total points** that **Michael Jordan** had during the 1995-96 regular season stored in the variable named "**jordan_points**":

```python
jordan_points = df.loc[df["Player"] == "Michael Jordan", "PTS"].values[0]
jordan_points
```

Now I am going to get the **Z-Score** which will represent how many **Standard Deviations** Michael Jordan is above the **mean** amount of points scored in the 1995-96 regular season:

```python
jordan_z_score = (jordan_points-mean_points)/std_points
jordan_z_score
```

Looking at the **Z-Score** calculated for **Michael Jordan's** total points scored in the regular season we get a **Z-score** of about 2.12, which(if we do the math) tells us that he scored about 1500 more points than the average from the players with more than 1000 minutes played. 

We can also just check that by subtracting the **mean_pts** from **Jordan's total points** as shown below:

```python
jordan_points - mean_points
```

Going back to that **Z-score** we calculated for Michael Jordan, it tells us nothing unless we compare it to the **Z-score** for points for other top players on the team:

I will be doing this by adding another column called **"points_zscore"** to the DataFrame **"df_over_1000"** to be able to easily compare the **Z-scores** of the players with over 1000 minutes played.

```python
df_over_1000["points_zscore"] = (df_over_1000["PTS"] - mean_points) / std_points
df_over_1000
```

By looking at the **"points_zscore"** column I just created we can easily see that **Jordan** has a significantly higher **z-score** (for points) when compared to all the other players that played over 1000 minutes.

He is only close competition is **Scottie Pippen** (who is regarded as an NBA Legend) who has a z-score for points that is about 0.74. Which is only about **540 points** above the mean total points, which does not even come close to Jordan's **1500 points** above the mean.

This analysis of points **(which is a very important statistic when comparing players)** already shows that Michael Jordan was **significantly better** than his teammates when looking at total points.

### Part 2.3: Calculating and Comparing Other Important Statistics Among Players with Over 1000 Minutes Played

Points aren't the only statistic that people look at and consider **important** when comparing players to each other.

People also look at statistics such as **Total Rebounds, Assists, Steals, and Blocks** to compare players.

In this part I will add more **z-score columns** (in the DataFrame **"df_over_1000"**) for each of these statistics to see which players are the best for each different type of statistic:

#### Adding a z-score column for **"Total Rebounds"** titled "rebounds_z_score" into DataFrame "df_over_1000":

```python
df_over_1000["rebounds_zscore"] = (df_over_1000["TRB"] - df["TRB"].mean()) / df["TRB"].std()
df_over_1000
```

#### Adding a z-score column for **"Assists"** titled "assists_z_score" into DataFrame "df_over_1000":

```python
df_over_1000["assists_zscore"] = (df_over_1000["AST"] - df["AST"].mean()) / df["AST"].std()
df_over_1000
```

#### Adding a z-score column for **"Steals"** titled "steals_z_score" into DataFrame "df_over_1000":

```python
df_over_1000["steals_zscore"] = (df_over_1000["STL"] - df["STL"].mean()) / df["STL"].std()
df_over_1000
```

#### Adding a z-score column for **"Blocks"** titled "blocks_z_score" into DataFrame "df_over_1000":

```python
df_over_1000["blocks_zscore"] = (df_over_1000["BLK"] - df["BLK"].mean()) / df["BLK"].std()
df_over_1000
```

#### Analysis of all new **"z-score"** columns:

When looking at the new columns added for the z-scores for **Points, Rebounds, Assists, Steals, and Blocks** there is a lot of information to take in all at once.

Just to reiterate these z-score values show how many  **standard deviations** the indiviual player is away from the **mean** value of the respective statistic **(specifically for the players with over 1000 minutes played)** For stats like: **Points, Rebounds, Assists, Steals, and Blocks**, a **Higher** and **Positive** z-score indicates that the respective player is above average in that statistic.

After taking a look at all of the new columns we can see that **Michael Jordan** has the highest z-score regarding **points**, **Dennis Rodman** has the highest z-score regarding **rebounds**, **Scottie Pippen** has the highest z-score regarding **assists**, **Michael Jordan** has the highest z-score regarding **steals**, and **Luc Longley** has the highest z-score regarding **blocks**.

That initial analysis of the z-scores may make it seem like **Michael Jordan** was only "good" in two categories(Points and Steals), and that it was **overall** a good team. But if you take a deeper look you can see that **Jordan** and **Pippen** as well, both of them had **positive z-scores** (with the exception of blocks for Jordan) for all of the z-score columns. That tells us that both of these players were above average in all of these statistics when compared to the players that played over 1000 minutes.

Although my initial observations and ideas for part two of the project would've been really good to answer my **initial question** which was to see whether or not **Michael Jordan** was **significantly better** than the rest of his teammates. I did some brainstorming and have decided to instead answer the question on whether **Michael Jordan** or **Scottie Pippen** was a more **impactful** player to the Bulls success that season.

In order to answer this question in Part 2 of my lab, I will be using **"Advanced"** Statistics data set from the 1995-1996 Regular Season and filtering them to only show rows for **Michael Jordan** and **Scottie Pippen** in order to try and answer my question.

### Part 5.1: Creating a new DataFrame for the **Advanced** Statistics:

**"Advanced"** Statistics in the basketball world are different from **"Normal" Box Score** Statistics because **Advanced** Statistics use complex math to make **"different"** types of statistics that show a player's overall impact to their teams, efficiency on the court, and value beyond regular stats.

Overall, **Advanced** Statistics are important because they measure true skill and team contribution istead of looking a basic numbers player's put up during a game. Essentially focusing on efficiency and skill.

#### "Advanced" Statistics Dataframe:

```python
df_advanced = pd.read_csv("Bulls_Advanced.csv")
df_advanced
```

### Filter the DataFrame in order to only have the players **Michael Jordan** and **Scottie Pippen**:

```python
dropped_rows = df_advanced.tail(14).index
advanced_m_s = df_advanced.drop(dropped_rows)
advanced_m_s
```

#### Initial Analysis of DataFrame:
When looking at **Advanced** Statistics in Basketball the most important statistics are **Player Efficiency Rating (PER)**, **Win Shares(WS)**, **Usage Percentage(USG%)**, and **True Shooting Percentage(TS%)**.

These **Advanced** Statistics will be used with a method of K-Means Clustering in order to see if the **clusters** for **Michael Jordan** and **Scottie Pippen** are significantly different, which can provide us with evidence to answer our question on whether **Michael Jordan** or **Scottie Pippen** contributed more to the 1995-1996 Chicago Bulls success.

### Part 5.2: Initialize the Machine Learning Model and Selecting Variables

#### Initializing:

```python
from sklearn.cluster import KMeans
kmeans = KMeans(2)
```

#### Selecting Variables:

```python
X = advanced_m_s[["PER", "WS", "USG%", "TS%"]]
```

#### Fit The Model:

```python
advanced_m_s["Cluster"] = kmeans.fit_predict(X)
advanced_m_s
```

### Part 5.3: Create a Scatterplot for KMeans Clustering:

For this clustering I will only be using the 2 most important **Advanced** Statistics in this situation, which are **Player Efficiency Rating(PER)** and **Win Shares(WS)**.

I will be using **PER** since that is a **measure of a players per-minute production that accounts for all stats** so it will make it more effective in comparing the two players. I will be using **WS** because that statistic is a measure of the **number of wins contributed by a player**. Win Shares is important to answer my main question because it can help us see who helped gain more wins for the team that season.

```python
import matplotlib.pyplot as plt
plt.scatter(advanced_m_s["PER"], advanced_m_s["WS"], c = advanced_m_s["Cluster"])
plt.xlabel("PER")
plt.ylabel("WS")
plt.title("Clustering of Jordan and Pippen using PER and WS")
plt.show()
```

### Part 5.4: Analysis of Clustering/Scatterplot:
Although it may not look like much in the scatterplot above, it actually tells us a lot. The two clusters representing **Michael Jordan** and **Scottie Pippen** are completely and distinctly separated with **Michael Jordan's** cluster having the better stats for **PER** and **WS**. The fact that these clusters are separated and that **Michael Jordan's** cluster clearly has better statistics for those two very important **"Advanced"** Statistics provides evidence on how **Michael Jordan** contributed significantly more to the team's success than **Scottie Pippen**.
